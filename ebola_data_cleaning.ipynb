```python
# libraries -----------------------------------------------------------------

import requests 
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# get data ----------------------------------------------------

df = pd.read_csv('ebola_data_db_format.csv', 
                 parse_dates=['Date'])
df.head()
```

```python
df.shape
```

```python
df['Indicator'].value_counts()
```

```python
df = df[(df['Indicator']=='Cumulative number of confirmed, probable and suspected Ebola cases')|
        (df['Indicator']=='Cumulative number of confirmed, probable and suspected Ebola deaths')|
        (df['Indicator']=='Cumulative number of suspected Ebola cases')|
        (df['Indicator']=='Cumulative number of suspected Ebola deaths')|
        (df['Indicator']=='Cumulative number of probable Ebola cases')|
        (df['Indicator']=='Cumulative number of probable Ebola deaths')|
        (df['Indicator']=='Cumulative number of confirmed Ebola cases')|
        (df['Indicator']=='Cumulative number of confirmed Ebola deaths')].reset_index(drop=True)
df.head()
```

```python
final_df = df.pivot_table(values='value', index=['Country', 'Date'], columns=['Indicator'])
final_df = final_df.reset_index()
final_df = final_df.sort_values('Date').reset_index(drop=True)
final_df.rename_axis(None)
final_df.head()
```

```python
final_df.columns
```

```python
final_df = final_df[['Country', 'Date', 'Cumulative number of suspected Ebola cases', 
                     'Cumulative number of probable Ebola cases', 'Cumulative number of confirmed Ebola cases', 
                     'Cumulative number of confirmed, probable and suspected Ebola cases',
                     'Cumulative number of suspected Ebola deaths', 'Cumulative number of probable Ebola deaths', 
                     'Cumulative number of confirmed Ebola deaths', 
                     'Cumulative number of confirmed, probable and suspected Ebola deaths']]
```

```python
final_df.columns = ['Country', 'Date', 'No. of suspected cases', 
                     'No. of probable cases', 'No. of confirmed cases', 
                     'No. of confirmed, probable and suspected cases',
                     'No. of suspected deaths', 'No. of probable deaths', 
                     'No. of confirmed deaths', 
                     'No. of confirmed, probable and suspected deaths']
```

```python
final_df.head()
```

```python
final_df['Country'].value_counts().sort_index()
```

```python
final_df['Country'] = final_df['Country'].replace(r'Liberia 2', 'Liberia', regex=True)
final_df['Country'] = final_df['Country'].replace(r'Guinea 2', 'Liberia', regex=True)
```

```python
final_df.to_csv('ebola_2014_2016_clean.csv', index=False)
```
